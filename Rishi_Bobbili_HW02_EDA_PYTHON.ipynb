```python


import pandas as pd
import matplotlib.pyplot as plt

```

```python
# I. Replace the file path with your actual path
file_path = "D:/Stevens/KDD/breast-cancer-wisconsin.csv"

# Reading the CSV file
data = pd.read_csv(file_path, header=None)

```

```python
# II. Summarize each column
print(data.describe())

```

```python
# III. Identify missing values
missing_values = data.isna().sum().sum()
print("Total missing values:", missing_values)


```

```python
# IV. Replace missing values with the mean of each column

# Calculate means for numeric columns
numeric_means = data.mean(numeric_only=True)

# Replace missing values with column-specific means
data = data.fillna(numeric_means)
```

```python

# V. Display scatter plots of F1 to F6 (one pair at a time)
for i in range(1, 6):
    for j in range(i + 1, 7):
        plt.scatter(data[i], data[j])
        plt.xlabel("F" + str(i))
        plt.ylabel("F" + str(j))
        plt.title("Scatter Plot of F{} vs. F{}".format(i, j))
        plt.show()

```

```python
# VI. Show histogram and box plots for columns F7 to F9
# Convert columns V7, V8, and V9 to numeric if they are not already
data[7] = pd.to_numeric(data[7], errors='coerce')
data[8] = pd.to_numeric(data[8], errors='coerce')
data[9] = pd.to_numeric(data[9], errors='coerce')

# Create histograms and box plots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))
data[7].plot(kind='hist', ax=axes[0, 0], title="Histogram of V7")
data[8].plot(kind='hist', ax=axes[0, 1], title="Histogram of V8")
data[9].plot(kind='hist', ax=axes[0, 2], title="Histogram of V9")
data[7].plot(kind='box', ax=axes[1, 0], title="Boxplot of V7")
data[8].plot(kind='box', ax=axes[1, 1], title="Boxplot of V8")
data[9].plot(kind='box', ax=axes[1, 2], title="Boxplot of V9")
plt.tight_layout()

# Display the resulting dataset after removing rows with any missing value in any column
data.dropna(inplace=True)
print(data)
```
